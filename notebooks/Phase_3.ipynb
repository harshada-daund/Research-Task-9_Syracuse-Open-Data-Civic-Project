## Exploratory Data Analysis (EDA)

```python
df.shape
```

```python
df.isna().sum().sort_values(ascending=False)
```

```python
violations_per_year = df.groupby('Year')['ObjectId'].count()
violations_per_year
```

```python
violations_per_year.plot.bar(figsize=(7,4))
plt.title('Housing Code Violations per Year')
plt.ylabel('Count')
plt.show()
```

```python
df.groupby('Is_Open')['ObjectId'].count()
```

```python
df['Violation_Clean'].value_counts().head(10)
```

```python
closed = df[(~df['Is_Open']) & (df['Resolution_Days'] >= 0)]
closed['Resolution_Days'].describe()
```

```python
closed['Resolution_Days'].plot.hist(bins=25, edgecolor='black', figsize=(7,5))
plt.title('Resolution Time Distribution (Closed Cases)')
plt.xlabel('Resolution Days')
plt.show()
```

### EDA Summary
- Violations vary by year with reporting lag in recent periods
- Many cases remain open, limiting duration analysis
- A small number of violation categories account for most records
- All findings are descriptive and subject to data limitations

## Expanded Exploratory Data Analysis (EDA)

```python
print(f"Total records: {len(df)}")
df.dtypes
```

```python
missing = df.isna().sum().sort_values(ascending=False)
missing
```

```python
missing[missing > 0].plot.bar(figsize=(8,4), title='Missing Values by Column')
plt.ylabel('Missing Count')
plt.show()
```

```python
violations_year = df.groupby('Year')['ObjectId'].count()
violations_year
```

```python
violations_year.plot.bar(figsize=(7,4), title='Housing Code Violations per Year')
plt.ylabel('Number of Violations')
plt.show()
```

```python
status_counts = df['Is_Open'].value_counts().rename({True:'Open', False:'Closed'})
status_counts
```

```python
status_counts.plot.bar(figsize=(6,4), title='Open vs Closed Violations')
plt.ylabel('Count')
plt.show()
```

```python
top_violations = df['Violation_Clean'].value_counts().head(10)
top_violations
```

```python
top_violations.sort_values().plot.barh(figsize=(8,5), title='Top 10 Violation Types')
plt.xlabel('Number of Violations')
plt.show()
```

```python
neighborhood_counts = df['Neighborhood_Clean'].dropna().value_counts().head(10)
neighborhood_counts
```

```python
neighborhood_counts.plot.bar(figsize=(8,4), title='Top 10 Neighborhoods by Violations')
plt.ylabel('Number of Violations')
plt.show()
```

```python
closed = df[(~df['Is_Open']) & (df['Resolution_Days'].notna()) & (df['Resolution_Days'] >= 0)]
print(f"Closed cases with valid resolution days: {len(closed)}")
if closed.empty:
    print('No closed cases with valid resolution days available.')
else:
    closed['Resolution_Days'].plot.hist(bins=20, edgecolor='black', figsize=(7,5))
    plt.title('Resolution Time Distribution (Closed Cases)')
    plt.xlabel('Resolution Days')
    plt.ylabel('Frequency')
    plt.show()
```
